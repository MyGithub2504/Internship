```python
# lets now import all the required libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
import warnings
warnings.filterwarnings("ignore")
```

```python
#Lets first connect to the web driver
driver = webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
driver.get(" https://www.trak.in")
```

```python
fund_btn=driver.find_element_by_xpath("//i[@class='bf-icon  fa fa-dollar']")
fund_btn.click()
```

```python
SR_NO=[]
Date=[]
Name=[]
Industry_Vertical=[]
Sub_vertical=[]
City_Location=[]
Investors_Name=[]
Investment_type=[]
Amount_USD=[]
```

```python
date=driver.find_elements_by_xpath("//td[@class='column-2']")
date[:]
```

```python
for i in date:
    dates=i.text
    Date.append(dates)
Date[:]
```

```python
srno=driver.find_elements_by_xpath("//td[@class='column-1']")
srno[:]
```

```python
for i in srno:
    srnos=i.text
    SR_NO.append(srnos)
SR_NO[:]
```

```python
name=driver.find_elements_by_xpath("//a[@data-wpel-link='external']")
name[:]
```

```python
for i in name:
    names=i.text
    Name.append(names)
Name[:]
```

```python
industry_vertical=driver.find_elements_by_xpath("//td[@class='column-4']")
industry_vertical[:]
```

```python
for i in industry_vertical:
    industry=i.text
    Industry_Vertical.append(industry)
Industry_Vertical[:]
```

```python
sub_vertical=driver.find_elements_by_xpath("//td[@class='column-5']")
sub_vertical[:]
```

```python
for i in sub_vertical:
    sub=i.text
    Sub_vertical.append(sub)
Sub_vertical[:]
```

```python
City=driver.find_elements_by_xpath("//td[@class='column-6']")
City[:]
```

```python
for i in City:
    city=i.text
    City_Location.append(city)
City_Location[:]
```

```python
Investors=driver.find_elements_by_xpath("//td[@class='column-7']")
Investors[:]
```

```python
for i in Investors:
    investor=i.text
    Investors_Name.append(investor)
Investors_Name[:]
```

```python
investment=driver.find_elements_by_xpath("//td[@class='column-8']")
investment[:]
```

```python
for i in investment:
    investments=i.text
    Investment_type.append(investments)
Investment_type[:]
```

```python
Amount_usd=driver.find_elements_by_xpath("//td[@class='column-9']")
Amount_usd[:]
```

```python
for i in Amount_usd:
    usd=i.text
    Amount_USD.append(usd)
Amount_USD[:]
```

```python
Funding=pd.DataFrame({})
Funding['Sr_No']=SR_NO[55:85]
Funding['Date']=Date[55:85]
Funding['Name']=Name[55:85]
Funding['Industry/Vertical']=Industry_Vertical[55:85]
Funding['Sub_Vertical']=Sub_vertical[55:85]
Funding['City_Location']=City_Location[55:85]
Funding['Investors_Name']=Investors_Name[55:85]
Funding['Investment_Type']=Investment_type[55:85]
Funding['Amount_USD']=Amount_USD[55:85]
Funding
```
