```python
# lets now import all the required libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
import warnings
warnings.filterwarnings("ignore")
```

```python
#Lets first connect to the web driver
driver = webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
driver.get("https://www.forbes.com/billionaires/")
```

```python
Rank=[]
Name=[]
Net_Worth=[]
Age=[]
Citizenship=[]
Source=[]
Industry=[]
```

```python
name=driver.find_elements_by_xpath("//div[@class='personName']")
name[0:190]
```

```python
for i in name:
    names=i.text
    Name.append(names)
Name[0:190]
```

```python
rank=driver.find_elements_by_xpath("//div[@class='rank']")
rank[0:190]
```

```python
for i in rank:
    ranks=i.text
    Rank.append(ranks)
Rank[0:190]
```

```python
networth=driver.find_elements_by_xpath("//div[@class='netWorth']")
networth[0:190]
```

```python
for i in networth:
    networths=i.text
    Net_Worth.append(networths)
Net_Worth[0:190]
```

```python
age=driver.find_elements_by_xpath("//div[@class='age']")
age[0:190]
```

```python
for i in age:
    ages=i.text
    Age.append(ages)
Age[0:190]
```

```python
citizenship=driver.find_elements_by_xpath("//div[@class='countryOfCitizenship']")
citizenship[0:190]
```

```python
for i in citizenship:
    cities=i.text
    Citizenship.append(cities)
Citizenship[0:190]
```

```python
source=driver.find_elements_by_xpath("//div[@class='source-column']")
source[0:190]
```

```python
for i in source:
    sources=i.text
    Source.append(sources)
Source[0:190]
```

```python
industry=driver.find_elements_by_xpath("//div[@class='category']")
industry[0:190]
```

```python
for i in industry:
    industries=i.text
    Industry.append(industries)
Industry[0:190]
```

```python
# import pandas as pd
Forbes=pd.DataFrame({})
Forbes['Ranks']=Rank[0:190]
Forbes['Names']=Name[0:190]
Forbes['Net_Worth']=Net_Worth[0:190]
Forbes['Age']=Age[0:190]
Forbes['Citizenship']=Citizenship[0:190]
Forbes['Source']=Source[0:190]
Forbes['Industry']=Industry[0:190]
Forbes
```
