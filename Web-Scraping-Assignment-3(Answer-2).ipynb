```python
Brand_name=[]
Product_name=[]
Ratings=[]
Number_of_Ratings=[]
Price=[]
Return_Exchange=[]
Expected_delivery=[]
```

```python
#Fetching urls
urls1=[]
for i in driver.find_elements_by_xpath("//a[@class='a-link-normal s-no-outline']"):
    urls1.append(i.get_attribute("href"))
```

```python
click_btn1=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[2]/div[1]/div[1]/div[1]/span[3]/div[2]/div[67]/span[1]/div[1]/div[1]/ul[1]/li[3]/a[1]")
click_btn1.click()
```

```python
#Fetching urls
urls2=[]
for i in driver.find_elements_by_xpath("//a[@class='a-link-normal s-no-outline']"):
    urls2.append(i.get_attribute("href"))
```

```python
urls2
```

```python
url=urls1+urls2
```

```python
for i in url:
    driver.get(i)
    time.sleep(3)
    #fetching brand
    try:
        brand=driver.find_element_by_xpath("/html[1]/body[1]/div[2]/div[2]/div[5]/div[3]/div[4]/div[41]/div[1]/table[1]/tbody[1]/tr[1]/td[2]/span[1]")
        Brand_name.append(brand.text.replace("/n","new line"))
    except :
        Brand_name.append('-')
        
    #Fetching prod_name 
    try:
        prod_name=driver.find_element_by_xpath("/html[1]/body[1]/div[2]/div[2]/div[5]/div[3]/div[4]/div[1]/div[1]/h1[1]/span[1]")
        Product_name.append(prod_name.text)
    except:
        Product_name.append('-')
         
    #Fetching Rating
    try:
        rating=driver.find_element_by_xpath("/html[1]/body[1]/div[4]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/span[1]")
        Ratings.append(rating.text)
    except:
        Ratings.append('-') 
      
 
     #Fetching Number_of_Ratings
    try:
        numrating=driver.find_element_by_xpath("/html[1]/body[1]/div[2]/div[2]/div[5]/div[3]/div[4]/div[5]/div[1]/span[3]/a[1]/span[1]")
        Number_of_Ratings.append(numrating.text)
    except:
        Number_of_Ratings.append('-') 
        
        #Fetching Price
    try:
        price=driver.find_element_by_xpath("/html[1]/body[1]/div[2]/div[2]/div[5]/div[3]/div[4]/div[9]/div[1]/div[1]/table[1]/tbody[1]/tr[2]/td[2]/span[1]")
        Price.append(price.text)
    except:
        Price.append('-') 
        
        #Fetching Return_exchange
    try:
        return_exchange=driver.find_element_by_xpath("/html[1]/body[1]/div[2]/div[2]/div[5]/div[3]/div[4]/div[20]/div[2]/div[1]/div[1]/div[1]/div[1]/span[1]/div[2]/a[1]")
        Return_Excahnge.append(return_exchange.text)
    except:
        Return_Exchange.append('-') 
        
        #Fetching Expected_delivery
    try:
        delivery=driver.find_element_by_xpath("/html[1]/body[1]/div[2]/div[2]/div[5]/div[3]/div[4]/div[12]/div[1]/div[1]/b[1]")
        Expected_delivery.append(delivery.text)
    except:
        Expected_delivery.append('-')  
        
```
