```python
# lets now import all the required libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
import warnings
from time import sleep
warnings.filterwarnings("ignore")
```

```python
#Lets first connect to the web driver
driver = webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/")
```

```python
btn1=driver.find_element_by_xpath("/html[1]/body[1]/div[3]/div[3]/div[5]/div[1]/table[1]/tbody[1]/tr[1]/td[1]/span[1]/a[1]")
btn1.click()
```

```python
btn2=driver.find_element_by_xpath("//body/div[@id='content']/div[@id='bodyContent']/div[@id='mw-content-text']/div[3]/ul[1]/li[1]/div[1]/a[1]/span[6]")
btn2.click()
sleep(5)
```

```python
driver.maximize_window()
```

```python
Serial_Number=[]
Video_Name=[]
Artist=[]
Views_billion=[]
Upload_Date=[]
```

```python
Serial_number=driver.find_elements_by_xpath('/html[1]/body[1]/div[3]/div[3]/div[5]/div[1]/table[3]/tbody[1]/tr/td[1]')
for sno in Serial_number:
    try:
        Serial_Number.append(sno.text)
    except NoSuchElementException:
        Serial_Number.append('-')
Serial_Number[0:30]
```

```python
Video_name=driver.find_elements_by_xpath('//body[1]/div[3]/div[3]/div[5]/div[1]/table[3]/tbody[1]/tr/td[2]')
for video in Video_name:
    try:
        Video_Name.append(video.text)
    except NoSuchElementException:
        Video_Name.append('-')
Video_Name[0:30]
```

```python
artist=driver.find_elements_by_xpath('/html[1]/body[1]/div[3]/div[3]/div[5]/div[1]/table[3]/tbody[1]/tr/td[3]')
for art in artist:
    try:
        Artist.append(art.text)
    except NoSuchElementException:
        Artist.append('-')
Artist[0:30]
```

```python
views=driver.find_elements_by_xpath('/html[1]/body[1]/div[3]/div[3]/div[5]/div[1]/table[3]/tbody[1]/tr/td[4]')
for view in views:
    try:
        Views_billion.append(view.text)
    except NoSuchElementException:
        Views_billion.append('-')
Views_billion[0:30]
```

```python
Upload_date=driver.find_elements_by_xpath('/html[1]/body[1]/div[3]/div[3]/div[5]/div[1]/table[3]/tbody[1]/tr/td[5]')
for upload in Upload_date:
    try:
        Upload_Date.append(upload.text)
    except NoSuchElementException:
        Upload_Date.append('-')
Upload_Date[0:30]
```

```python
print(len(Serial_Number),len(Video_Name),len(Artist),len(Views_billion),len(Upload_Date))
```

```python
Youtube=pd.DataFrame({})
Youtube['Serial_number']=Serial_Number[0:30]
Youtube['Video_name']=Video_Name[0:30]
Youtube['Artist']=Artist[0:30]
Youtube['Views_Billion']=Views_billion[0:30]
Youtube['Upload_Date']=Upload_Date [0:30]
Youtube
```
