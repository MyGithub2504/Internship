```python
# lets now import all the required libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
import warnings
import os
from time import sleep
warnings.filterwarnings("ignore")
```

```python
#Lets first connect to the web driver
driver = webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
driver.get("https://www.imdb.com/list/ls095964455/")
```

```python
Name=[]
Year_Span=[]
Genre=[]
Run_Time=[]
Ratings=[]
Votes=[]
```

```python
name=driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a")
for i in name:
    try:
        Name.append(i.text)
    except NoSuchElementException:
        Name.append('-')
Name[0:100]
```

```python
year=driver.find_elements_by_xpath("//h3[@class='lister-item-header']/span")
for i in year:
    try:
        Year_Span.append(i.text)
    except NoSuchElementException:
        Year_Span.append('-')
Year_Span[0:100]
```

```python
Year_Span=Year_Span[1:200:2]
Year_Span
```

```python
genre=driver.find_elements_by_xpath("//div[@class='lister-item-content']/p[1]/span[5]")
for i in genre:
    try:
        Genre.append(i.text)
    except NoSuchElementException:
        Genre.append('-')
Genre[0:100]
```

```python
runtime=driver.find_elements_by_xpath("//div[@class='lister-item-content']/p[1]/span[3]")
for i in runtime:
    try:
        Run_Time.append(i.text)
    except NoSuchElementException:
        Run_Time.append('-')
Run_Time[0:100]
```

```python
rating=driver.find_elements_by_xpath("//div[@class='ipl-rating-widget']/div[1]/span[2]")
for i in rating:
    try:
        Ratings.append(i.text)
    except NoSuchElementException:
        Ratings.append('-')
Ratings[0:100]
```

```python
votes=driver.find_elements_by_xpath("//p[@class='text-muted text-small']/span[2]")
for i in votes:
    try:
        Votes.append(i.text.replace('|',""))
    except NoSuchElementException:
        Votes.append('-')
Votes[0:100]
```

```python
Votes=Votes[1:200:2]
Votes
```

```python
print(len(Name),len(Year_Span),len(Genre),len(Run_Time),len(Ratings),len(Votes))
```

```python
IMDB=pd.DataFrame({})
IMDB["Name"]=Name
IMDB["Year-Span"]=Year_Span
IMDB["Genre"]=Genre
IMDB["Run-Time"]=Run_Time
IMDB["Ratings"]=Ratings
IMDB["Votes"]=Votes
IMDB
```
