```python
title=[]
ratings=[]
price=[]
```

```python
#Fetching urls
urls=[]
for i in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
    urls.append(i.get_attribute("href"))
```

```python
urls
```

```python
for i in urls:
    driver.get(i)
    time.sleep(3)
    #fetching title
    try:
        titles=driver.find_element_by_xpath("/html[1]/body[1]/div[2]/div[2]/div[5]/div[3]/div[4]/div[1]/div[1]/h1[1]/span[1]")
        title.append(titles.text.replace("/n","new line"))
    except :
        title.append('-')
        
    #Fetching ratings 
    try:
        Ratings=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[2]/div[1]/div[1]/div[1]/span[3]/div[2]/div[2]/div[1]/span[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[2]/div[1]/div[1]/div[2]/div[1]/span[1]/span[1]/a[1]/i[1]")
    except:
        ratings.append('-')
         
    #Fetching price
    try:
        Price=driver.find_element_by_xpath("/html[1]/body[1]/div[2]/div[2]/div[5]/div[3]/div[4]/div[9]/div[1]/div[1]/table[1]/tbody[1]/tr[2]/td[2]/span[1]")
        price.append(Price.text)
    except:
        price.append('-') 
```

```python
# import pandas as pd
laptop=pd.DataFrame({})
laptop['title']=title[0:10]
laptop['Ratings']=ratings[0:10]
laptop['Price']=price[0:10]
```
