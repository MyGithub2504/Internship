```python
soup10=BeautifulSoup(page.content)
soup10
```

```python
print(soup10.prettify)
```

```python
housetitle=soup10.find_all('h2', class_="heading-6 font-semi-bold nb__1AShY")
housetitle
```

```python
housetitles = [] #empty list for store 
for i in housetitle:
    housetitles.append(i.text.replace('\n',''))
housetitles
```

```python
houselocation=soup10.find_all('div', class_="nb__35Ol7")
houselocation
```

```python
houselocations = [] #empty list for store 
for i in houselocation:
    houselocations.append(i.text.replace('\n',''))
houselocations
```

```python
threeparameters=soup10.find_all('div', class_="font-semi-bold heading-6")
threeparameters
```

```python
threeparameter = [] #empty list for store 
for i in threeparameters:
    threeparameter.append(i.text.replace('\n',''))
threeparameter
```

```python
housearea=[]
for point in threeparameter[0::3]:
    housearea.append(point)
housearea
```

```python
emiofhouse=[]
for point in threeparameter[1::3]:
    emiofhouse.append(point)
emiofhouse
```

```python
houseprice=[]
for point in threeparameter[2::3]:
    houseprice.append(point)
houseprice
```

```python
import pandas as pd
NOBroker=pd.DataFrame({})
NOBroker['Housetitles']=housetitles
NOBroker['Houselocations']=houselocations
NOBroker['HouseArea']=housearea
NOBroker['EMIof house']=emiofhouse
NOBroker['HousePrice']=houseprice
NOBroker
```
