```python
brand=[]
product_desc=[]
price=[]
discount=[]
```

```python
for i in urlcomplete:
    driver.get(i)
    time.sleep(3)
    #fetching brand
    try:
        bd=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[1]/div[3]/div[1]/div[2]/div[2]/div[1]/div[1]/h1[1]/span[1]")
        brand.append(bd.text.replace("/n","new line"))
    except :
        brand.append('-')
        
        
    #Fetching product description
    #prodcutdesc_btn=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[1]/div[3]/div[1]/div[2]/div[7]/div[3]/div[1]/div[1]/div[1]/div[1]")
    #productdesc_btn.click()
    try:
        pdc=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[1]/div[3]/div[1]/div[2]/div[7]/div[3]/div[1]/div[1]/div[2]/div[1]/div[1]")
        product_desc.append(pdc.text)
    except:
        product_desc.append('-')
        
      #Fetching price
    try:
        pr=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[1]/div[3]/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div[1]/div[1]")
        price.append(pr.text)
    except:
        price.append('-')  
        
       #Fetching discount 
    try:
        disc=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[1]/div[3]/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div[1]/div[3]/span[1]")
        discount.append(disc.text)
    except:
        discount.append('-')  
          
```
