```python
# lets now import all the required libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
import warnings
import os
from time import sleep
warnings.filterwarnings("ignore")
```

```python
#Lets first connect to the web driver
driver = webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
driver.get("https://www.billboard.com/")
```

```python
btn=driver.find_element_by_xpath("//header/div[1]/ul[1]/li[1]/a[1]")
btn.click()
```

```python
btn1=driver.find_element_by_xpath("//body/main[@id='main']/div[5]/div[1]/div[2]/div[2]/a[1]/div[1]/div[1]")
btn1.click()
```

```python
Song_Name=[]
Artist_Name=[]
Last_week_rank=[]
Peak_rank=[]
Weeks_on_board=[]
```

```python
song_name=driver.find_elements_by_xpath("//span[@class='chart-element__information']/span[1]")
for i in song_name:
    try:
        Song_Name.append(i.text)
    except NoSuchElementException:
        Song_Name.append('-')
Song_Name[0:100]
```

```python
artist=driver.find_elements_by_xpath("//span[@class='chart-element__information']/span[2]")
for j in artist:
    try:
        Artist_Name.append(j.text)
    except NoSuchElementException:
        Artist_Name.append('-')
Artist_Name[0:100]
```

```python
rank=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
for k in rank:
    try:
        Last_week_rank.append(k.text)
    except NoSuchElementException:
        Last_week_rank.append('-')
Last_week_rank[0:100]
```

```python
peak=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
for r in peak:
    try:
        Peak_rank.append(r.text)
    except NoSuchElementException:
        Peak_rank.append('-')
Peak_rank[0:100]
```

```python
weeks=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
for q in weeks:
    try:
        Weeks_on_board.append(q.text)
    except NoSuchElementException:
        Weeks_on_board.append('-')
Weeks_on_board[0:100]
```

```python
print(len(Song_Name),len(Artist_Name),len(Last_week_rank),len(Peak_rank),len(Weeks_on_board))
```

```python
Billboard=pd.DataFrame({})
Billboard["Song_Name"]=Song_Name[0:100]
Billboard["Artist_Name"]=Artist_Name[0:100]
Billboard["Last_Week_Rank"]=Last_week_rank[0:100]
Billboard["Peak_Rank"]=Peak_rank[0:100]
Billboard["Weeks_On_Board"]=Weeks_on_board[0:100]
Billboard
```
