```python
Brand=[]
Prod_desc=[]
Price=[]
Discount=[]
```

```python
for i in url:
    driver.get(i)
    time.sleep(3)
    #fetching brand
    try:
        brand=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[1]/div[3]/div[1]/div[2]/div[2]/div[1]/div[1]/h1[1]/span[1]")
        Brand.append(brand.text.replace("/n","new line"))
    except :
        Brand.append('-')
        
    #Fetching prod_desc 
    try:
        prod_desc=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[1]/div[3]/div[1]/div[2]/div[2]/div[1]/div[1]/h1[1]/span[2]")
        Prod_desc.append(prod_desc.text)
    except:
        Prod_desc.append('-')
         
    #Fetching price
    try:
        price=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[1]/div[3]/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div[1]/div[1]")
        Price.append(price.text)
    except:
        Price.append('-') 
      
    #Fetching discount
    try:
        discount=driver.find_element_by_xpath("/html[1]/body[1]/div[1]/div[1]/div[3]/div[1]/div[2]/div[2]/div[1]/div[3]/div[1]/div[1]/div[3]/span[1]")
        Discount.append(discount.text)
    except:
        Discount.append('-') 
```
