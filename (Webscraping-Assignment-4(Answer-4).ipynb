```python
# lets now import all the required libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
import warnings
import os
from time import sleep
warnings.filterwarnings("ignore")
```

```python
#Lets first connect to the web driver
driver = webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
driver.get("https://www.statisticstimes.com/economy/india-statistics.php")
```

```python
btn=driver.find_element_by_xpath("/html[1]/body[1]/div[2]/div[2]/div[2]/ul[1]/li[1]/a[1]")
btn.click()
```

```python
Rank=[]
State=[]
GSDP_19_20=[]
GSDP_18_19=[]
Share_18_19=[]
GDP_Billion=[]
```

```python
rank=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[1]")
for j in rank:
    try:
        Rank.append(j.text)
    except:
        Rank.append('-')
        Rank.append('NoSuchElementException')
Rank[0:]
```

```python
state=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[2]")
for j in state:
    try:
        State.append(j.text)
    except:
        State.append('-')
        State.append('NoSuchElementException')
State[0:]
```

```python
gsdp_19_20=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[3]")
for k in gsdp_19_20:
    try:
        GSDP_19_20.append(k.text)
    except:
        GSDP_19_20.append('-')
GSDP_19_20[0:]
```

```python
gsdp_18_19=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[4]")
for l in gsdp_18_19:
    try:
        GSDP_18_19.append(l.text)
    except:
        GSDP_18_19.append('-')
        GSDP_18_19.append('NoSuchElementException')
GSDP_18_19[0:]
```

```python
share_18_19=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[5]")
for m in share_18_19:
    try:
        Share_18_19.append(m.text)
    except:
        Share_18_19.append('-')
        Share_18_19.append('NoSuchElementException')
Share_18_19[0:]
```

```python
gdp_billion=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[6]")
for n in gdp_billion:
    try:
        GDP_Billion.append(n.text)
    except:
        GDP_Billion.append('-')
        GDP_Billion.append('NoSuchElementException')
GDP_Billion[0:]
```

```python
print(len(Rank),len(State),len(GSDP_19_20),len(GSDP_18_19),len(Share_18_19),len(GDP_Billion))
```

```python
Stats=pd.DataFrame({})
Stats["Rank"]=Rank
Stats["State"]=State
Stats["GSDP_19_20"]=GSDP_19_20
Stats["GSDP_18_19"]=GSDP_18_19
Stats["Share_18_19"]=Share_18_19
Stats["GDP_Billion"]=GDP_Billion
Stats
```
